```python
#imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# The file location
merged_stats = "https://raw.githubusercontent.com/csc442-team11/MLB-Data-Analysis/refs/heads/main/Datasets/mlb-player-stats-Batters-2023-2024-cleaned.csv"


# Read in the file
df_merged = pd.read_csv(merged_stats)
```

games played on position and at bats


```python
#scatterplot of games and at bats
sns.scatterplot(x="G", y="AB", data=df_merged);
```

```python
#barplot of games and position
sns.barplot(x="G", y="Pos", data=df_merged);
```

```python
#barplot of at bats and position
sns.barplot(x="AB", y="Pos", data=df_merged);
```

```python
#crosstabulation of teams and positions
pd.crosstab(df_merged["Team"], df_merged["Pos"])
```

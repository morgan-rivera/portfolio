```python
# Import necessary libraries
import requests as rq
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# Try and retrieve the HTML from the website
def get_HTML(url):
    try:
        r = rq.get(url)
    except rq.ConnectionError:
        print("Connection Error")
        return None
    except rq.HTTPError:
        print("HTTP Error")
        return None
    except rq.TimeoutError:
        print("Timeout Error")
        return None
    return bs(r.content, "html.parser")
```

```python
# Retrieve the tables from the HTML
def get_tables(soup):
    roster_tables = soup.find_all("table", {"class": "roster__table"})
    roster_tables = roster_tables[2:]
    return roster_tables
```

```python
# Retrieve the players from the specific tables
def parse_table(table, players):
    body = table.find("tbody")
    rows = body.find_all("tr")
    for row in rows:
        player = row.find("td", {"class": "info"}).find("a").get_text()
        players.append(player)
```

```python
def get_players(team):
  # The URL of the website to scrap
  roster_url = "https://www.mlb.com/" + team + "/roster/depth-chart"
  # gets HTML from the url
  soup = get_HTML(roster_url)
  # breaks HTML into a list of tables (where the players are stored)
  roster_tables = get_tables(soup)

  players = []
  for table in roster_tables:
    # Finds all the players from each table
    parse_table(table, players)

  return list(set(players))
```
